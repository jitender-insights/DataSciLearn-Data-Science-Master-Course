# Seaborn 

- Seaborn is a Python data visualization library built on top of Matplotlib. It is particularly useful for creating statistical graphics and provides functions for easily visualizing complex datasets.


```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
#df['sepal_length']
df.sepal_length
```

```python
sns.scatterplot(x=df.sepal_length , y = df.sepal_width)
```

```python
sns.scatterplot(x=df.sepal_length , y = df.petal_length)
```

```python
sns.scatterplot(x=df.petal_width , y = df.petal_length)
```

```python
sns.displot(df['sepal_length'])
```
