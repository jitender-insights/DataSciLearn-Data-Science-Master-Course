# Bokeh     

- Bokeh is another Python library for creating interactive and web-based visualizations. It is well-suited for creating interactive, data-driven applications and dashboards.

```python
pip install bokeh
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure , output_file , show
from bokeh.sampledata.iris import flowers

```

```python
flowers
```

```python
output_file('sample.html')
p = figure(title='Flower Data Set')
p.xaxis.axis_label ='patel_length'
p.yaxis.axis_label ='patel_width'
p.scatter(flowers['petal_length'], flowers['petal_width'])
show(p)
```

```python
output_file('sample.html')
p = figure(title='Flower Data Set')
p.xaxis.axis_label ='patel_length'
p.yaxis.axis_label ='patel_width'
p.line(flowers['petal_length'], flowers['petal_width'])
show(p)
```

```python
x = [2,5,6,8,9]
y = [3,4,5,6,7]
output_file('line.html')
p = figure(title = "Line Plot")
p.line(x,y)
show(p)
```

```python
x = [2,5,6,8,9]
y = [3,4,5,6,7]
output_file('circle.html')
p = figure(title = "Scatter Plot")
p.circle(x,y , fill_color='red' , size=30 )
show(p)
```
