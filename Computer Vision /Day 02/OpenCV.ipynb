# 1. Reading and displaying images

```python
import cv2
import matplotlib.pyplot as plt

# Read the image from the file 

image = cv2.imread("bird.jpg")

# Display the image
cv2.imshow('Image',image)
# plt.imshow(image)

# Wait for a key press and close the image window
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 2. Converting images to grayscale

```python
# Conert the image to grayscale

gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

# Display the image
cv2.imshow('Grayscale Image',gray_image)

# Wait for a key press and close the image window
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 3. Edge Detection Using Canny

```python
# Perform Canny edge detection

edges = cv2.Canny(gray_image,100,200)
# The two arguments (100 and 200) are the threshold values for edge detection



# Display the image
cv2.imshow('Edges',edges)

# Wait for a key press and close the image window
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 4. Image Filtering and Edge Detection

```python
import matplotlib.pyplot as plt
# Gaussian Blur
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
#Kernel Size (5, 5) and Standard Deviation 0

# Canny Edge Detection
edges = cv2.Canny(blurred_image, 100, 200)

# Display the edges
plt.imshow(edges, cmap='gray')
plt.show()
```

# Image Processing with OpenCV

```python
import cv2
import numpy as np

# Load an image
image = cv2.imread('street.jpg')

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Detect edges using Canny
edges = cv2.Canny(blurred_image, 100, 200)

# Show images
cv2.imshow('Original Image', image)
cv2.imshow('Gray Image', gray_image)
cv2.imshow('Blurred Image', blurred_image)
cv2.imshow('Edges', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Drawing shapes and text on images

```python
import numpy as np

# Draw a line
cv2.line(image, (0, 0), (511, 511), (255, 0, 0), 5)

# Draw a rectangle
cv2.rectangle(image, (384, 0), (510, 128), (0, 255, 0), 3)

# Draw a circle
cv2.circle(image, (447, 63), 63, (0, 0, 255), -1)

# Draw an ellipse
cv2.ellipse(image, (256, 256), (100, 50), 0, 0, 180, 255, -1)

# Draw a polygon
pts = np.array([[10, 5], [20, 30], [70, 20], [50, 10]], np.int32)
pts = pts.reshape((-1, 1, 2))
cv2.polylines(image, [pts], True, (0, 255, 255))

# Put text on the image
cv2.putText(image, 'OpenCV', (10, 500), cv2.FONT_HERSHEY_SIMPLEX, 4, (255, 255, 255), 2, cv2.LINE_AA)

# Display the image with shapes and text
cv2.imshow('Image with Shapes', image)

# Wait for a key press and close the image window
cv2.waitKey(0)
cv2.destroyAllWindows()
```
