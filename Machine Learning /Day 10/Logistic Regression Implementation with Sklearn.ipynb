## Logistic Regression Implementation

### Binary Classification

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
dataset = load_breast_cancer()
```

```python
print(dataset.DESCR)

#WDBC-Malignant ---Cancerous
#WDBC-Benign- Non_cancerous
```

```python
dataset.keys()
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df
```

```python
df['target'] = dataset.target
```

```python
df.head()
```

```python
df.tail()
```

# Independent and Dependent features

```python
X = df.iloc[:,:-1]
X
```

```python
y= df.iloc[:,-1]
y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test , y_train , y_test = train_test_split_test_split(X , y , test_size=0.20 , random_state=42)
```

```python
classifier.fit(X_train , y_train)
```

```python
classifier.predict_proba(X_test)
```

```python
# Prediction
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

## Confusion Matrix , accuracy score , classification report 

```python
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report
```

```python
print(confusion_matrix(y_test , y_pred))
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test , y_pred))
```

## Hyperparameter Tuning

```python
# Grid search CV
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
paramters = {'penalty':('l1','l2','elasticnet',None) , 'C':[1,10,20]}
```

```python
clgs = GridSearchCV(classifier, param_grid=paramters , cv=5)
```

### Split train data to validation data

```python
clgs.fit(X_train , y_train)
```

```python
clgs.best_params_
```

```python
classifier = LogisticRegression(C=20 , penalty='l2')
```

```python
classifier.fit(X_train , y_train)
```

```python
y_pred = classifier.predict(X_test)
print(confusion_matrix(y_test , y_pred))
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test , y_pred))
```
